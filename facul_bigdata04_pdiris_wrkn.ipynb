# **Tratamento de dados com biblioteca Pandas**

```python
import pandas as pd
import numpy
import plotly.express as px
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
iris_data_df = pd.read_csv(io.BytesIO(uploaded['iris.csv']))
```

```python
iris_data_df
```

```python
iris_data_df.columns
```

```python
iris_data_df.fillna(0.0, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
data = iris_data_df [[c for c in iris_data_df.columns if c !='variety']]
scaler = MinMaxScaler()
scaler.fit(data)
preprocessed_data_df = scaler.transform(data)
```

```python
preprocessed_data_df = pd.DataFrame(preprocessed_data_df, columns = [c for c in iris_data_df.columns if c !='variety'])
```

```python
preprocessed_data_df ['variety'] = iris_data_df['variety']
```

```python
preprocessed_data_df
```

```python
px.histogram(preprocessed_data_df, x="sepal.length", color="variety")
```

```python
## O arquivo pode ser baixado clicando na pastinha do lado no GUI (creio que seja o sample_data)
from google.colab import files
preprocessed_data_df.to_csv("new_iris_data.csv")
##files.download("new_iris_data.csv")
```
